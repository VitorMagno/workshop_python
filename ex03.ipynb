```python
import numpy as np
x = np.array ([i for i in range(0,101,1)])
```

```python
print(x)
```

```python
for i in x:
  y = x*x
```

```python
print(y)
```

```python
import pandas as pd
rng = np.random.default_rng(12345)
rnd_int = rng.integers(low = 0, high = 101, size = 101)
print (rnd_int)
```

```python
df = pd.DataFrame(rnd_int)
```

```python
df.to_csv("numbers.csv")
z = pd.read_csv("numbers.csv", index_col = False)
```
